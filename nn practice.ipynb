# 레모네이드 판매량 예측

```python
file = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/lemonade.csv'

import tensorflow as tf
import pandas as pd

# 파일 가져오기
lemon = pd.read_csv(file)
lemon.head()
```

```python
독립 = lemon[['온도']]
종속 = lemon[['판매량']]
print(독립.shape, 종속.shape)
```

```python
# 독립변수의 칼럼개수가 1이기에 shape = 1
x = tf.keras.layers.Input(shape=[1])
# 종속변수 칼럼개수가 1이기에 dense = 1
y = tf.keras.layers.Dense(1)(x)
model = tf.keras.models.Model(x,y)
# 모델 학습 방법
model.compile(loss='mse')
```

## 모델 학습

```python
# verbose = 화면 출력 조정(0 = 출력 없음)
model.fit(독립, 종속, epochs=1000, verbose = 0)
# 출력이 없으므로 확인하기 위해 10개만 출력
model.fit(독립, 종속, epochs=10)
model.predict(독립)
model.predict([[15]])
```

# 보스턴 집값 예측

```python
file = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/boston.csv'
bst = pd.read_csv(file)
print(bst.columns)
bst.head()
```

## 변수구성


```python
ind = bst.drop('medv', axis=1)
dep = bst[['medv']]

print(ind.shape, dep.shape)
```

## 모델 구성

```python
x = tf.keras.layers.Input(shape=[13])
y = tf.keras.layers.Dense(1, activation='softmax') (x) #softmax = 확률 추론(dense가 2이상의 경우, 확률)
model = tf.keras.models.Model(x,y)
model.compile(loss = 'mse')
```

## 모델 학습

```python
model.fit(ind, dep, epochs = 1000, verbose = 0)
model.fit(ind, dep, epochs=10)
```

```python
print(model.predict(ind[5:10]))
print(dep[5:10])

print(model.get_weights())
```
